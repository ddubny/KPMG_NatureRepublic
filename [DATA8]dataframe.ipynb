```python
a = df[0:20]
b = df[3887:3907]
df_chat = pd.concat([a,b])
```

```python
# 데이터 생성 및 불러오기 


# 문헌 분류값 리스트 불러오기

classify = pd.read_csv('/Users/Samsung/Documents/KPMG/classify_df.csv', encoding='utf-8', engine="python").set_index('Unnamed: 0')

# 담당사원 정보 랜덤으로 생성하기 

import random

emp = ['서수빈', '김명헌', '윤채원', '백지원', '임정민', '최낙현' ] 
emps = []

for i in range(len(df_chat)) : 
    x = random.choice(emp)
    emps.append(x)


# audit_dic = dict(적정= 4, 한정 = 3, 부적정 = 2, 의견거절 = 1)

audit_level = ['적정', '한정', '부적정', '의견거절'] 
audit_levels = []

for i in range(len(df_chat)) : 
    x = random.choice(audit_level)
    audit_levels.append(x) 

###### 딕셔너리값으로 변환하기 
audit_dic = dict(적정= 4, 한정 = 3, 부적정 = 2, 의견거절 = 1)
audit_score = []

for i in range(len(audit_levels)) : 
    y = audit_dic.get(audit_levels[i])
    audit_score.append(y)
    

# 회사 규모 데이터 임의 생성하기 


size = []

for i in range(len(df_chat)) : 
    x = random.randrange(100000, 99999999999)
    size.append(x) 
```

```python
pre = pd.read_excel('/Users/Samsung/Documents/KPMG/pre_train.xlsx')

pre_train = pre.drop(columns=['Unnamed: 0'], axis=1).set_index(['Unnamed: 1']).transpose()
pre_train_df = pre_train.rename(columns={pre_train.columns[0]:'corp_name'}).reset_index().drop(columns=['index']).set_index(['corp_name'])
pre_train_dic = pre_train_df.to_dict('index')

pre_train_dic
```

```python
pre_train_dic
#print(list(zip(pre_train_dic.keys(), pre_train_dic.values())))
name = list(pre_train_dic.keys()) 
info = list(pre_train_dic.values())
```

```python
df_chat['classify'] = classify
df_chat['manager'] = emps
df_chat['recent_progress'] = "재무제표 확인중"
df_chat['last_audit grade'] = audit_levels
df_chat['corp_size'] = size
df_chat['kye_list']


# 슬랙 데이터 임의로 추가해두기. 
df_chat['recent_progress'][2] = '담당자 변경 알림: 감사A팀 최낙현 → 감사B팀 김명헌' 
```
